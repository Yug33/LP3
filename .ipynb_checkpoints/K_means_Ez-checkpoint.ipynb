```python
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
x = np.array([[0.1,0.6],[0.15,0.71],[0.08,0.9],[0.16,0.85],[0.2,0.3],[0.25,0.5],[0.24,0.1],[0.3,0.2]])
```

```python
def find_distance(c,x):
    return np.sum(np.abs(c-x))
```

```python
c1=x[0]
c2=x[7]
c1=c1.reshape(1,2)
c2=c2.reshape(1,2)
```

```python
for i in range(20):
    cluster1=[]
    cluster2=[]
    for i in range(len(x)):
        print(x[i])
        d1=find_distance(x[i],c1)
        d2=find_distance(x[i],c2)
        if (d1<=d2):
            cluster1.append(x[i])
        else:
            cluster2.append(x[i])
        c1=np.mean(cluster1,axis=0)
        c2=np.mean(cluster2,axis=0)
```
